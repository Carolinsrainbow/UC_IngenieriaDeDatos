## 2. Esquema

Para esta actividad vamos a trabajar con datos de una empresa que vende productos a través de varias tiendas. Trabajaremos con el siguiente esquema:

- Tiendas(tid PRIMARY KEY, capacidad, minpersonal)
- Productos(pid PRIMARY KEY, nombre, precio, tipo)
- Vende(tid,pid) PRIMARY KEY(tid,pid)
- Usuarios(uid PRIMARY KEY, nombre, calle, comuna, region)
- Compras(cid PRIMARY KEY, uid, tid)
- Contenido_Compra(cid,pid,cantidad) PRIMARY KEY (cid,pid)

En este esquema, la tabla 'Tiendas' tiene información sobre cada tienda de la empresa. La información que guardaremos sobre cada tienda es su id único, la capacidad de personas que pueden ingresar a la tienda, y la cantidad mínima del personal requerido para el funcionamiento de la tienda. El repositorio (único) de productos se guarda en la tabla 'Productos', donde cada producto se describe a través de su id, nombre, su precio y tipo. Para registrar que una tienda en particular vende un producto, se ocupa la tabla 'Vende', vinculando la tienda con el producto. Por otro lado, la tabla 'Usuarios' tiene un registro de los usuarios de la empresa, respaldando la información con su nombre y dirección.

Para registrar las compras que los usuarios realizan de una tienda específica, se tienen dos tablas. La primera, llamada 'Compras' establece una compra de un usuario en una tienda, y la tabla 'Contenido_Compra' guarda la información sobre los productos comprados en cada compra individual. También se guarda la información sobre la cantidad de unidades de cada producto comprado para poder modelar que un usuario compra varias unidades del mismo producto en una compra.

## 4. Consultas [6 puntos]

Escriban las siguentes consultas SQL que funcionan sobre el esquema de arriba. Cada consulta vale 0.5 puntos.


**1**. Liste los productos desde el más caro hasta el más barato:

```python
%%sql
SELECT *
FROM Productos
ORDER BY precio DESC;
```

2. Muestre en una sola tabla la información de las tiendas y sus productos

```python
%%sql
SELECT *
FROM Tiendas, Vende, Productos
WHERE Tiendas.tid = Vende.tid
AND Vende.pid = Productos.pid;
```

A este tipo de consulta se le llama Join, consiste en unir la información de varias tablas, haciéndolas coincidir en ciertos valores (usualmente, con llaves). Se pueden hacer coincidir tantos valores como sean necesarios añadiendo más lineas con AND.

Existen varias formas de hacer Joins, esta es solo una de ellas.

3. Máximo de los precios para cada tipo de producto. Las respuestas deberían listar pares [tipo, máximo(precio)].

GROUP BY: sentencia mediante la cual decimos a la consulta que agrupe las tuplas según los valores de una columna

<img src='https://learnsql.com/blog/group-by-in-sql/v7.gif'>

Como imagen: https://drive.google.com/uc?id=139NNvTGetxI5QWyUO_uctDDOJv1FbhZ1

```python
%%sql
SELECT tipo, precio
FROM Productos
ORDER BY tipo;
```

```python
%%sql
SELECT tipo, MAX(precio)
FROM Productos
GROUP BY tipo
ORDER BY MAX(precio) DESC;
```

4. Tienda con el número de clientes que realizaron compras en esta tienda. Ojo que hay que eliminar usuarios duplicados.

```python
%%sql
SELECT tid AS tienda, uid
FROM Compras
ORDER BY tienda;
```

```python
%%sql
SELECT tid AS tienda, COUNT( DISTINCT uid) AS numero_de_clientes
FROM Compras
GROUP BY tienda;
```

5. Nombre del usuario, junto con el precio del producto más caro que
compró. Si el usuario no realizó ninguna compra, no es necesario listarlo:

```python
%%sql
SELECT *
FROM Usuarios, Compras, Contenido_Compra, Productos
WHERE Usuarios.uid = Compras.uid AND Compras.cid = Contenido_Compra.cid AND Contenido_Compra.pid = Productos.pid
```

```python
%%sql
SELECT Usuarios.nombre, MAX(Productos.precio) AS producto_mas_caro
FROM Usuarios, Compras, Contenido_Compra, Productos
WHERE Usuarios.uid = Compras.uid AND Compras.cid = Contenido_Compra.cid AND Contenido_Compra.pid = Productos.pid
GROUP BY Usuarios.nombre
ORDER BY producto_mas_caro DESC;
```

6. Id de cada tienda, junto con el promedio de los precios de todos los productos que vende esta tienda

```python
%%sql
SELECT tid As id_tienda, AVG(Productos.precio) AS precio_promedio
FROM Vende, Productos
WHERE Vende.pid = Productos.pid
GROUP BY tid;
```

7. Nombre de los Productos que contienen una B mayúscula:

8. Nombre de los Productos que comienzan con una B mayúscula:

```python
%%sql
SELECT *
FROM Productos AS p
WHERE p.nombre LIKE 'B%';
```

Basicamente, el % se puede interpretar como "cualquier texto", entonces:
* %B%: Una B mayúscula, y lo que sea a la izquierda y la derecha
* B%: Una B mayúscula, y lo que sea a la derecha
* %B: Lo que sea, que termine en B mayúscula
* B%B: Algo que inicie y termine con B mayúscula
